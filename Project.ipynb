```python
#Task 1
#There are three different tables as given above. Please make three dataframe in python
#and save them as three .csv files. From Task 2 to Task 10, use the saved .csv files only

#Created seniority_data dictionary of lists
#Each key is a column name, and the values are lists of column data.
#Id,Designation level are keys are keys in dictionary. Assigned  list values to keys


seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
}
#Create seniority data frame from dictionary of list
seniority_df = pd.DataFrame(seniority_data)
##Loading seniority dataframe to seniority.csv file
seniority_df.to_csv("seniority.csv", index=False)
```

```python
###Task 2
#compute project missing values. Replace the missing values by running average

##import pandas and numpy libraries
#Import the pandas library for data manipulation
import pandas as pd
#Import the numpy library for nan
import numpy as np

# Reload created project CSV and assign it to variable proj
proj = pd.read_csv("project.csv")
# Reload created employee CSV and assign it to variable  emp
emp = pd.read_csv("employee.csv")
# Reload created seniority CSV and assign it to variable sen
sen = pd.read_csv("seniority.csv")

# Introduce NaN values (simulate missing costs)
# Using loc[] is a label-based data selection method in pandas, used to access rows and columns by labels
#Set the value in row 5, column 'Cost' to NaN (missing)
proj.loc[4, "Cost"] = np.nan
#Set the value in row 9, column 'Cost' to NaN (missing)
proj.loc[8, "Cost"] = np.nan


#Convert 'Cost' column to a  Python list
costs = proj["Cost"].tolist()
#Initialize cumulative sum and count of valid (non-NaN) entries
running_sum, count = 0, 0
#########################
# Loop through each cost value by index
for i in range(len(costs)):
    if pd.isna(costs[i]):  # if cost is missing
        if count > 0:      # only replace if we already have some valid numbers
            avg = running_sum / count #compute running average from previous values
            costs[i] = avg #Replace NaN with the average
            running_sum += avg ## Include the new value in sum and count
            count += 1 #Increase count
    else:                  # valid number → include in running sum
        running_sum += costs[i] # # Add it to the running sum
        count += 1 ##Increase count
        # Replace the original 'Cost' column with the updated list 
proj["Cost"] = costs
#Print proj dataframe
print(proj,"\n")
```

```python
#Task 3
#Split the name column in the Employee dataframe into two new columns “First Name”,
#and “Last Name” and remove the older “name” column.


# Using split function to create 2 first name and last name
# Split the 'Name' column into two new columns: 'FirstName' and 'LastName'
# - str.split splits on the first space only
# - expand=True turns the result into a DataFrame (with two columns)

emp[["FirstName","LastName"]] = emp["Name"].str.split(" ", n=1, expand=True)
# Drop the original 'Name' column since we’ve now split it into two
# axis=1 means drop a column 
# inplace=True makes the change directly in the emp DataFrame
emp.drop("Name", axis=1, inplace=True)
#Print messgae
print("✅ Task 3: Name column split into FirstName and LastName\n")
#Prrint data frame
print(emp, "\n")
```

```python
#Task 4
#Join all three dataframes in one single dataframe. Name it “Final”
#Using merge function in python to join all dataframes.
#using id for merging
#Merge 'proj' and 'emp' on the 'ID' column 
# Merge the result with 'sen' on the 'ID' column
final = proj.merge(emp, on="ID").merge(sen, on="ID")
#Print message
print("✅ Task 4: Final DataFrame created\n")
#Print final data frame
print(final, "\n")
```

```python
#Task 5
#Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
#cost only to employees who have finished the projects
#final.apply applies the function row by row
#Calculate 5% of cost using row['cost']*0.05
#If row status is finished, give bonus, else no bonus.
final["Bonus"] = final.apply(lambda row: row["Cost"]*0.05 if row["Status"]=="Finished" else 0, axis=1)

print("✅ Task 5: Bonus column added\n")
print(final, "\n")
```

```python
#Task 6
#Demote the designation level by 1, whose projects have status “fail”. Delete the
#employees record whose designation level is above 4

# For all rows where 'Status' is 'Failed', increase the 'Designation Level' by 1
# This simulates demoting the employee if their project failed
final.loc[final["Status"]=="Failed","Designation Level"] += 1
# Keep only those rows where 'Designation Level' is less than or equal to 4
final = final[final["Designation Level"] <= 4]
# Print message
print("✅ Task 6: Demotions applied and >4 removed\n")
# Print final data frame
print(final, "\n")


```

```python

#Task 7
#Add “Mr.” and “Mrs.” to the first name column and drop the gender column.
#Using apply function row by row.
# Using apply with a lambda to process each row individually
final["FirstName"] = final.apply(lambda row: ("Mr." if row["Gender"] == "M" else "Mrs.") + row["FirstName"], axis=1)

# Remove the 'Gender' column from the DataFrame as it's no longer needed
final.drop("Gender", axis=1, inplace=True)
#Print message
print("✅ Task 7: Titles added and Gender dropped\n")
#Print final dataframe
print(final, "\n")
```

```python
#Task 8
#Promote designation level by 1 for the employees whose age is more than 29 years using
#IF condition.
#For each row, check if 'Age' is greater than 29
# If yes, subtract 1 from 'Designation Level' (promote)
# Otherwise, keep 'Designation Level' unchanged
final["Designation Level"]=final.apply(lambda row:row["Designation Level"] -1 if row["Age"]>29 else row["Designation Level"],axis=1)
#print message
print("✅ Task 8: Promotions applied for Age > 29\n")
#print final data frame
print(final, "\n")
```

```python
#Task 9
#Add the cost of all projects for each Employee and save it in new dataframe
#“TotalProjCost” with three columns ID, First Name, and Total cos
#Using group by ID & First name to find the sum of cost
#Groups projects by employee (ID and FirstName).

#Sums all project costs for each employee.
total_proj_cost = final.groupby(["ID","FirstName"])["Cost"].sum().reset_index()
# Rename the 'Cost' column to 'TotalCost' 
total_proj_cost.rename(columns={"Cost":"TotalCost"}, inplace=True)
#Load total_proj_cost dataframe to total_proj_cost.csv file
total_proj_cost.to_csv("total_proj_cost.csv", index=False)
#print message
print("✅ Task 9: Total Project Cost per Employee\n")
#print total_proj_cost data frame
print(total_proj_cost, "\n")
```

```python
#Task 10
#Print all the employee details whose city name contains the letter “o” in it
#Returns a Boolean Series: True where the city contains "o" or "O", False otherwise
#Result is stored in a new DataFrame: employees_with_o
employees_with_o = final[final["City"].str.contains("o", case=False)]
#Print message
print("✅ Task 10: Employees in cities containing 'o'\n")
#print employees_with_o
print(employees_with_o, "\n")
```
